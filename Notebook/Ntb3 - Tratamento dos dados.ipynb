# Valores omissos

```python
#FUNÇÃO PARA IDENTIFICAR VALORES OMISSOS
def mostra_missing_value(df):
    total = df.isnull().sum().sort_values(ascending = False)
    percent = 100*round((df.isnull().sum()/df.isnull().count()).sort_values(ascending = False), 2)
    missing_data = pd.concat([total, percent], axis = 1, keys=['Total', 'Percentual'])
    print(missing_data.head(15))
```

```python
#QUANTIDADE DE VALORES OMISSOS
mostra_missing_value(df_target)
```

### Tratamento para padrões temporais

```python
df=df_target.groupby(['País', 'Ano']).agg({'Tax_%_GDP': ['mean'],'GINI_Index': ['mean'], 
                                           'Social_benefits_expense_%_GDP': ['mean'], 'Interest_expense_%_GDP': ['mean'],
                                           'OilRrents_%_GDP': ['mean'], 'Agriculture_%_GDP': ['mean'], 
                                           'ControlofCorruption': ['mean'],'GovernmentEffectiveness':['mean'], 
                                           'RegulatoryQuality': ['mean'], 'RuleofLaw': ['mean'], 
                                           'Political_StabilityNoViolence': ['mean'], 'VoiceandAccountability': ['mean']})
```

```python
df2 = df.groupby("País").transform(lambda x: x.ffill())
```

```python
df3 = df2.groupby("País").transform(lambda x: x.bfill())
```

```python
mostra_missing_value(df3)
```

```python
mostra_missing_value(df3)
```

### Tratamento para imputação de 0 (zero)

```python
df3['OilRrents_%_GDP'].fillna(0, inplace=True)
df3['Interest_expense_%_GDP'].fillna(0, inplace=True) 
df3['Social_benefits_expense_%_GDP'].fillna(0, inplace=True) 
df3['Agriculture_%_GDP'].fillna(0, inplace=True) 
```

```python
mostra_missing_value(df3)
```

### Tratamento para imputação por meio do algoritmo dos K vizinhos mais próximos

```python
df_base_line1 = df3.copy()
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
```

```python
df4 = pd.DataFrame(imputer.fit_transform(df_base_line1), columns = df_base_line1.columns, index = df_base_line1.index)
```

```python
mostra_missing_value(df4)
```
