# Pré-processamento Carga Tributária

```python
df_cargaTributária = pd.read_csv(pastaTrab + '\\FMI\\GFSR_02-04-2022 20-03-01-71_timeSeries.csv')
```

```python
df_cargaTributária.shape
```

```python
df_cargaTributária.info()
```

```python
df_cargaTributária.head()
```

```python
df_cargaTributária.columns
```

```python
df_boll = df_cargaTributária['Attribute'] == 'Value'
df_cargaTributária_filtrado = df_cargaTributária[df_boll].drop(columns = ['Country Code', 'Classification Name',
                                                                                'Classification Code', 'Sector Name', 'Sector Code', 'Unit Name',
                                                                                'Unit Code', 'Attribute', 'Indicator Code',
                                                                                'Global DSD Time Series Code'])
df_cargaTributária_filtrado.rename(columns = {'Country Name': 'Country'}, inplace = True)
```

```python
columns = ['2001', '2002', '2003', '2004', '2005','2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014','2015',
           '2016', '2017', '2018', '2019', '2020'] 
df_cargaTributária_filtrado[columns] = df_cargaTributária_filtrado[columns].apply(lambda x: x.str.replace(',', '.').astype('float'))
```

```python
df_cargaTributária = df_cargaTributária_filtrado.groupby(['Country']).sum()
df_cargaTributária
```

```python
df_cargaTributária['Indicador'] = 'FiscalEffort'
df_cargaTributária
```

```python
df_cargaTributária.to_excel(pastaTrab + '\\FMI\\cargaTributaria_py.xlsx')
```

# Pré-processamento WWGI

```python
df_WWGI1 = pd.read_excel(pastaTrab + '\\WWGI\\wgidataset.xlsx', sheet_name = 'VoiceandAccountability', skiprows=12)
df_WWGI1['Indicador'] = 'VoiceandAccountability'
df_WWGI2 = pd.read_excel(pastaTrab + '\\WWGI\\wgidataset.xlsx', sheet_name = 'ControlofCorruption', skiprows=12)
df_WWGI2['Indicador'] = 'ControlofCorruption'
df_WWGI3 = pd.read_excel(pastaTrab + '\\WWGI\\wgidataset.xlsx', sheet_name = 'RuleofLaw', skiprows=12)
df_WWGI3['Indicador'] = 'RuleofLaw'
df_WWGI4 = pd.read_excel(pastaTrab + '\\WWGI\\wgidataset.xlsx', sheet_name = 'RegulatoryQuality', skiprows=12)
df_WWGI4['Indicador'] = 'RegulatoryQuality'
df_WWGI5 = pd.read_excel(pastaTrab + '\\WWGI\\wgidataset.xlsx', sheet_name = 'GovernmentEffectiveness', skiprows=12)
df_WWGI5['Indicador'] = 'GovernmentEffectiveness'
df_WWGI6 = pd.read_excel(pastaTrab + '\\WWGI\\wgidataset.xlsx', sheet_name = 'Political StabilityNoViolence', skiprows=12)
df_WWGI6['Indicador'] = 'Political_StabilityNoViolence'
```

```python
df_boll_null = df_WWGI_filter['Unnamed: 0'].notnull()
df_WWGI1 = df_WWGI_filter[df_boll_null]
df_boll_c = df_WWGI1['Unnamed: 0'] != 'Country/Territory'
df_WWGI_filt = df_WWGI1[df_boll_c]
```
